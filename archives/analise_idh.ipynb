## Organizando Dados

```python
features = mun_2010.iloc[:,5:] # indicadores de IDH
```

```python
# Ordenando siglas manualmente
auxs = np.array(['RO','AC','AM','RR','PA','AP','TO','MA','PI', 'CE','RN','PB','PE','AL','SE','BA',
        'MG','ES','RJ','SP','PR','SC','RS','MS','MT','GO','DF'])

# Agrupando códigos, nomes e siglas de UF's
uf_ = np.array([pd.unique(code['UF']),pd.unique(code['Nome_UF']),auxs]).T
uf_
```

```python
ss = []
uf_[:,[0,2]]

for i in range(len(mun_2010)):
    for ii in range(len(uf_)):
        if mun_2010.loc[i,'UF']==uf_[ii,0]:
            ss.append(uf_[ii,2])    
```

```python
print(len(ss))
print(len(mun_2010))
code['siglas']=ss
```

```python
code
```

```python
from unidecode import unidecode

aux_ep = []
aux_index = []
for i in range(len(code)):
    for ii in range(len(mun)):
        if [mun.loc[ii,'UF'],unidecode(mun.loc[ii,'NOME DO MUNICIPIO'])]==[code.loc[i,'siglas'],unidecode(code.loc[i,'Nome_Munic'].upper())]:
            aux_index.append(i)
            aux_ep.append(mun.loc[ii,'ESPECTRO'])
            continue
                                      
```

```python
print(code.iloc[aux_index,:].shape)
print(features.iloc[aux_index,:].shape)
print(len(aux_ep))
```

```python
code = code.iloc[aux_index,:]
features = features.iloc[aux_index,:]
code['espectro'] = aux_ep
```
