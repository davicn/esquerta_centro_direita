# Análise Quantitativa - Municipios

```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt 
%matplotlib notebook
import seaborn as sns

path = os.getcwd().replace('archives','')
path
```

```python
data = pd.read_csv(path + '/docs/result_data.csv')
data
```

```python
l = ['D','D','E','E','E','E','E','D','E','C','D','D','E','D','D','C','D','E','E','E','D','D','E','E']
ll = pd.unique(data['SIGLA DO PARTIDO'])
lll = []
for i in range(len(ll)):
    lll.append([ll[i] ,l[i]])
lll = np.array(lll)
tipos_partidos = pd.DataFrame(data=lll,columns=['PARTIDO','ESPECTRO'])
tipos_partidos.head()
```

```python
aux_partidos = []

for i in range(len(data)):
    for ii in range(len(tipos_partidos)):
        if data.loc[i,'SIGLA DO PARTIDO']==tipos_partidos.loc[ii,'PARTIDO']:
            aux_partidos.append(tipos_partidos.loc[ii,'ESPECTRO'])
            
aux_partidos
```

```python
data['ESPECTRO'] = aux_partidos
data
```

```python
data[(data['CODIGO DO MUNICIPIO']==1392) & (data['ESPECTRO']=='D')].loc[:,'NUM FILIADOS'].sum()
```

```python
municipios = data.loc[:,['UF','CODIGO DO MUNICIPIO','NOME DO MUNICIPIO']].drop_duplicates()
municipios.index = np.arange(len(municipios))
municipios
```

```python
def whatSpec(mun):
    aux = []
    e = ['D','E','C']
    for i in e:
        aux.append(data[(data['CODIGO DO MUNICIPIO']==mun[1]) & (data['ESPECTRO']==i)].loc[:,'NUM FILIADOS'].sum())
    return e[np.argmax(aux)]

aux_mun = []
for i in range(len(municipios)):
    mun = municipios.iloc[i].to_numpy()
    aux_mun.append(whatSpec(mun))
```

```python
len(aux_mun)
```

```python
municipios['ESPECTRO'] = aux_mun
municipios
```

```python
municipios.to_csv('municipios.csv')
```
